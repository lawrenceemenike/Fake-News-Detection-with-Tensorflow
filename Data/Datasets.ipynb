### Import all the needed packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import sklearn.datasets
import tensorflow as tf
import tensorflow_datasets as tfds
```

### Let's load a few datasets from those available in sklearn.datasets package

```python
digits = sklearn.datasets.load_digits()
```

```python
# Handwritten Digit Dataset Visualization

figure = plt.figure(figsize=(12,6)) # matplotlib figure
for i in range(1,11):
    image = np.array(digits['images'][i], dtype="float") # convert to
                                                       # NumPy format for visualization
    figure.add_subplot(2,5,i)  # add a subplot to matplotlib figure
    plt.imshow(image, cmap='Greys')  # visualize greyscale image
plt.show()  # show all
```

```python
# Understanding the structure of obtained data

print(str(digits['target_names'].shape) + "- possible labels used in the dataset")
print(str(digits['target'].shape) + "- labels of entries in the dataset")
print(str(digits['images'].shape) + "- images in the dataset")
```

```python
# Regression dataset
X, y = sklearn.datasets.load_boston(return_X_y=True) # return_X_y means 2 separate
                    # dictionaries are returned for data and labels
```

```python
# Housing Prices Dataset Visualization

figure = plt.figure(figsize = (16,16))
for i in range(X.shape[1]): # for all features of X 
    ax = figure.add_subplot(4,4,i+1) # add a subplot
    ax.set_ylabel('price')
    ax.set_xlabel('feature' + str(i+1))
    plt.scatter(X[:,i], y) # visualize particular X feature
plt.show()
# Data may be used for predicting housing prices
```

```python
#Load Iris Flower database from tensorflow-datasets

tfds.disable_progress_bar() # might cause issues with dependencies
my_data = tfds.load(name='iris', batch_size=-1)
```

```python
# Convert data from Tensorflow tensors to NumPy arrays

numpy_ds = tfds.as_numpy(my_data)
numpy_data, numpy_labels = numpy_ds['train']['features'], numpy_ds['train']["label"] 
# separate features and labels
```

```python
#Visualize Iris Flower database

figure = plt.figure(figsize=(16,8))
colors =['red','green', 'blue', 'black'] # use different colors for different features
for i in range(numpy_data[0].size): # all features in dataset
    ax = figure.add_subplot(4,4,i+1) # add a subplot
    ax.set_ylabel('Flower type') # y-axis label is flower type
    ax.set_xlabel('feature' + str(i+1)) # x-axis is a particular feature
    plt.scatter(numpy_data[:,i], numpy_labels, color=colors[i]) # visualize
plt.show()
```
